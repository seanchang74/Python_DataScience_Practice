### Ufuncs & Aggregations

```python
theta = np.linspace(0,np.pi,3)
print(theta)
print(np.sin(theta))
print(np.cos(theta))
print(np.tan(theta))
```

```python
x = np.arange(5)
y = np.empty(5)
np.multiply(x,10,out=y)
y
```

```python
x = np.arange(1,6)
np.add.reduce(x)
```

```python
np.add.accumulate(x)
```

```python
np.multiply.outer(x,x)
```

```python
M = np.random.random((3,4))
M
```

```python
M.min(axis=0)
```

```python
M.max(axis=1)
```

```python
import pandas as pd
data = pd.read_csv('dataset/president_heights.csv')
heights = np.array(data['height(cm)'])
print(heights)
```

```python
print('Mean height: ',heights.mean())
print('Standard deviation: ',heights.std())
print('Minimum height: ',heights.min())
print('Maximum height: ',heights.max())
```

```python
print('25th percentile: ',np.percentile(heights,25))
print('Median: ',np.median(heights))
print('75th percentile: ',np.percentile(heights,75))
```

```python
import matplotlib.pyplot as plt
plt.hist(heights)
plt.title('Height Distribution of US Presidents')
plt.xlabel('height(cm)')
plt.ylabel('number')
```

<a id='Fancy_Indexing_Sort'></a>
### Fancy Indexing & Sorting

```python
rand = np.random.RandomState(42)
x = rand.randint(100, size=10)
print(x)
```

```python
ind = [3,7,4]
x[ind]
```

```python
x = np.zeros(10)
x[[0,0]] = [4,6]
print(x)
```

```python
i = [2,3,3,4,4,4]
x[i] += 1
x
```

```python
x = np.zeros(10)
np.add.at(x,i,1)
print(x)
```

```python
x = np.array([5,4,3,2,1])
np.sort(x)
```

```python
i = np.argsort(x)
print(i)
```

<a id='Numpy_Structured_Arrays'></a>
### Numpy's Structured Arrays

```python
name = ['Alice','Bob','Cathy','Doug']
age = [25,45,37,19]
weight = [55.0,85.5,68.0,61.5]
```

```python
data = np.zeros(4,dtype={'names':('name','age','weight'),'formats':('U10','i4','f8')})
print(data.dtype)
```

```python
data['name'] = name
data['age'] = age
data['weight'] = weight
print(data)
```

```python
data['name']
```

```python
data[0]
```

```python
data[data['age'] < 30]['name']
```
